8. Create a Boolean mask that is `True` when `vehicle_type` is `suv` or `pickup`. Filter `vehicles` down to only the `True` rows.

```python
suv = vehicles['vehicle_type'] == 'suv'
pickup = vehicles['vehicle_type'] == 'pickup'

suv_pickup = suv | pickup

vehicles[suv_pickup]
```

9. Create a Boolean mask that is `True` when `vehicle_type` is neither `suv` nor `pickup`. Filter `vehicles` down to only the `True` rows.

```python
not_suv_pickup = ~suv_pickup

vehicles[not_suv_pickup]
```

10. Create a Boolean mask that is `True` whenever `fuel` is `hybrid electric` and `best_mpge` is below average (less than 40). Filter `vehicles` down to only the rows that are `True`.

```python
hybrid_electric = vehicles['fuel'] == 'hybrid electric'
mpge_b40 = vehicles['best_mpge'] < 40

combined = hybrid_electric & mpge_b40

vehicles[combined]
```
